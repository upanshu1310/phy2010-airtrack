```python
names = {
    't': 'Latest: Time (s)',
    's1':'Latest: GateState 1',
    's2':'Latest: GateState 2'
}
```

```python
# df
```

```python
colors = ['red','blue','green','orange','purple','black']

for i in range(1,7):
    df = pd.read_csv(f'../data/partA/try{i}.csv')

    l = 1e-2
    
    dt1 = df[names['t']][1] - df[names['t']][0]
    dt2 = df[names['t']][3] - df[names['t']][2]
    dt3 = df[names['t']][5] - df[names['t']][4]
    dt4 = df[names['t']][7] - df[names['t']][6]

    v1 = l/dt1
    v2 = l/dt2
    v3 = l/dt3
    v4 = l/dt4

    plt.scatter(i,v1,color=colors[0],label='photogate 2')
    # plt.scatter(i,v2,color=colors[i-1])
    plt.scatter(i,v3,color=colors[1],label='photogate 1')
    # plt.scatter(i,v4,color=colors[i-1])

    abs_error = abs(v3-v1)
    rel_error = 100*2*abs_error/(v1+v2)

    print(f'{i}, {round(abs_error,2)}, {round(rel_error,2)}%')

plt.xlabel('iteration')
plt.ylabel('v (m/s)')
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())
plt.show()
```
